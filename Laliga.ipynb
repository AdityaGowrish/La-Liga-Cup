Print all the teams which have started playing between 1930-1980. (5 points)

```python
df.dtypes
```

```python
i = 0
year = []
size = df['Debut'].size
while(i < size):
    val = df['Debut'].values[i]
    years = val.split("-")
    year.append(years[0])
    i = i + 1
    
df['StartYear'] = year
start_year = df['StartYear'] >= '1930'
end_year = df['StartYear'] <= '1980'

print("Teams which have started playing between 1930-1980 are:\n{}".format(df[start_year & end_year]['Team']))
```

Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
#Goal_diff_count = GoalsFor - GoalsAgainst
df['GoalsFor'] = pd.to_numeric(df['GoalsFor'])
df['GoalsAgainst'] = pd.to_numeric(df['GoalsAgainst'])
goal_diff_count = lambda row_df : (row_df['GoalsFor'] - row_df['GoalsAgainst'])
df['GoalsDiffCount'] = df.apply(goal_diff_count, axis = 1)
print("Teams and their Goals Difference Counts are:\n{}".format(df[['Team','GoalsDiffCount']]))
```

```python
#Find the team which has maximum Goals Differences
df[df['GoalsDiffCount'] == df['GoalsDiffCount'].max()][['Team','GoalsDiffCount']]
```

```python
#Find the team which has minimum Goals Differences
df[df['GoalsDiffCount'] == df['GoalsDiffCount'].min()][['Team','GoalsDiffCount']]
```

Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'])
df['GamesWon'] = pd.to_numeric(df['GamesWon'])

def percentage_of_winning(row_df):
    percentage = 0
    try:
       percentage = (row_df['GamesWon'] / row_df['GamesPlayed']) * 100
    except ZeroDivisionError:
       persentage = pd.Series()
    return percentage.fillna(0).round(2)
    
df['WinningPercent'] = percentage_of_winning(df)
print("#The Top 5 teams which has the highest Winning percentage are:\n{}".format(df.sort_values(by='WinningPercent', ascending=False).head(5)[['Team','WinningPercent']]))
```
